```python
import pandas as pd

df = pd.read_csv("education_career_success.csv")

df.head()
```

```python
df.shape
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Job_Offers"], bins=10, kde=True)
plt.title("Job Offers Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Starting_Salary"], bins=30, kde=True)
plt.title("Salary Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x="University_GPA",
    y="Starting_Salary",
    data=df
)
plt.title("GPA vs Salary")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x="Years_to_Promotion",
    y="Starting_Salary",
    data=df
)
plt.title("Promotion Time vs Salary")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y="Field_of_Study",
    data=df,
    order=df["Field_of_Study"].value_counts().index
)
plt.title("Field of Study Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Career_Satisfaction"], bins=20, kde=True)
plt.title("Career Satisfaction Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x="Internships_Completed",
    y="Starting_Salary",
    data=df
)
plt.title("Internships vs Salary")
plt.show()
```

```python
print("Top Insights:")
print("- Average Salary:", df["Starting_Salary"].mean())
print("- Average GPA:", df["University_GPA"].mean())
print("- Average Satisfaction:", df["Career_Satisfaction"].mean())
print("- Most Common Field:", df["Field_of_Study"].mode()[0])
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Starting_Salary"])
plt.title("Salary Outliers")
plt.show()
```
